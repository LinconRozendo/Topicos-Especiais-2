```python
import matplotlib.pyplot as plt


x = []
y = []

for i in range(0, 11):
  x.append(i**2)
  y.append(i)

plt.plot(x,y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gr√°fico')
plt.savefig('fig.png')
  
```
