```python
import random # é necessário importar a biblioteca random para gerar os números aleatórios

cond = True
while (cond == True):
  tamanho = int(input('Digite o tamanho da lista, entre 0 a 50: '))
  if tamanho < 0 or tamanho > 50:
    print('Número digitado fora do intervalo permitido.')
  else:
    cond = False

lista = []

for i in range(0, tamanho):
  lista.append(random.randint(0, 10)) # o intervalo de 0 a 50 foi tomado por suposição

print(lista)

cond = True
while (cond == True):
  numero = int(input('Digite um número: '))

  cond_num = True
  while (cond_num == True):
    if numero in lista:
      lista.remove(numero)
    elif numero == 0:
      cond = False
      cond_num = False
    else:
      cond_num = False

print(lista)
```

```python
tamanho = int(input('Digite o tamanho da lista: '))

lista = []
lista_removida = []

for i in range(0, tamanho):
  lista.append(int(input('Digite um valor qualquer: ')))


cond = True
cont = 0
while (cond == True):
  if cont == len(lista):
    cond = False
    break

  if lista.count(lista[cont]) > 1:
    aux = lista[cont]
    lista.reverse()
    cond2 = True
    while (cond2 == True):
      if lista.count(aux) == 1:
        cond2 = False
        cont = 0
        lista.reverse()
      else:
        lista_removida.append(aux)
        lista.remove(aux)  
  else:
    cont += 1


print(lista)
print(lista_removida)
```
