1) Escreva um programa que leia dois números inteiros Lin e Col, que representam, respectivamente, a quantidade de linhas e colunas em uma matriz. Utilizando listas aninhas, crie uma representação para essa matriz, utilizando a função randint parar gerar números para cada posição da matriz. Apresente-a na tela com uma aparência matricial.

```python
import random

lin = int(input('Digite a quantidade de linha(s) da matriz: '))
col = int(input('Digite a quantidade de colunas(s) da matriz: '))
m = []

for i in range(0, lin):
  m.append([])
  
  for j in range(0, col):
    m[i].append(random.randint(0, 100))

for i in range(0, lin):

  for j in range(0, col):
    print(m[i][j], end = ' ')
    
  print(end = '\n')
```

2) Escreva um programa que leia do teclado duas matrizes de dimensões 2 x 2 e mostre na tela a soma dessas duas matrizes.

```python
m1 = []
m2 = []

for i in range(0, 2):
  m1.append([])

  for j in range(0, 2):
    numero = int(input('Digite qualquer valor para a matriz: '))
    m1[i].append(numero)

for i in range(0, 2):
  m2.append([])

  for j in range(0, 2):
    numero = int(input('Digite qualquer valor para a matriz: '))
    m2[i].append(numero)

soma = []

for i in range(0, 2):
  soma.append([])

  for j in range(0, 2):
    soma[i].append(m1[i][j] + m2[i][j])


for i in range(0, 2):

  for j in range(0, 2):
    print(soma[i][j], end = ' ')
    
  print(end = '\n')
```
