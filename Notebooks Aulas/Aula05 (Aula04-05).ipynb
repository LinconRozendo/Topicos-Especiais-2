# Gr√°fico de Barra

```python
nomes = ['Carros', 'Motos', 'Bicicletas']
quantidades = [50, 100, 200]
plt.bar(nomes, quantidades)
plt.show()
```

# Gr√°fico de Pizza

```python
nomes = ['Carros', 'Motos', 'Bicicletas']
quantidades = [10, 30, 20]
plt.pie(quantidades, labels = nomes, autopct = '%1f%%')
plt.show()
```
